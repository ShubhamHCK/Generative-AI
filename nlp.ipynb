NLP PHASE -1 : TEXT PROCESSING

```python
# Text processing pipeline
# text-> cleaning->tokenization->stopwards removal-> steeming/lemetation
```

```python
text = "i am learning Python programming and python is fast !!.."
print("Original text:",text)
```

```python
# step1 convert to lower case
text=text.lower()
print("LowerCase:",text)
```

```python
# step2 remove punctuation
# re - regex - regular expression
import re
text = re.sub(r'[^\w\s]','',text)
#remove everything except letter and spaces
print("without punctuation",text)
```

```python
#stpe3 Tokenization (splits sentences into words)
tokens=text.split()
print("Tokens;",tokens)
```

```python
#Step4 Remove stopwords - like is am are etc
stopword=["a","am","is","i","are","the","and"]
filtered_tokens=[]

for words in tokens:
    if words not in stopword:
        filtered_tokens.append(words)

print(" After Stopwords removal:",filtered_tokens)
```

```python
#stemming
# stemming
def stemp(word):
  if word.endswith("ing"):
    return word[:-3]
  elif word.endswith("ed"):
    return word[:-2]
  else:
      return word

stemmed_words=[]
for words in filtered_tokens:
  stemmed_words.append(stemp(words))

print("After Stemming:",stemmed_words)
```

NLP PHASE-2 WORD TO VECTOR

```python
# will convert text -> count
from sklearn.feature_extraction.text import CountVectorizer
```

```python
sentences =[
    "i am learninig python",
    "we are happy today",
    "i am sad today"
]
```

```python
vectorizer = CountVectorizer()  # create object
X= vectorizer.fit_transform(sentences)
# learns vocabulary(gives the different words)
# convret sentences to numbers
# print(X.toarray())
```

```python
print("Vocabulary:",vectorizer.get_feature_names_out())
#1 all unique words
#2 sorted in alphabetical order
#3 remove single letters like  i,a,
```

```python
print("Vectors:")
print(X.toarray())
```
